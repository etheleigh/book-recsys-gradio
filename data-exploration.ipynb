```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
def dataframe_statistics(df):
    stats = pd.DataFrame({
        'Column': df.columns,
        'Data Type': df.dtypes.values,
        'Non-Null Count': df.notnull().sum().values,
        'Missing Count': df.isnull().sum().values,
        'Missing %': (df.isnull().sum() / len(df) * 100).values,
        'Unique Values': df.nunique().values,
        'Min': [df[col].min() if np.issubdtype(df[col].dtype, np.number) else np.nan for col in df.columns],
        'Max': [df[col].max() if np.issubdtype(df[col].dtype, np.number) else np.nan for col in df.columns],
        'Mean': [df[col].mean() if np.issubdtype(df[col].dtype, np.number) else np.nan for col in df.columns],
        'Median': [df[col].median() if np.issubdtype(df[col].dtype, np.number) else np.nan for col in df.columns],
        'Std Dev': [df[col].std() if np.issubdtype(df[col].dtype, np.number) else np.nan for col in df.columns]
    })
    
    return stats
```

```python
dataframe_statistics(books)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax=plt.axes()

sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)
```

```python
books['missing description'] = np.where(books["description"].isna(), 1, 0)
books[('age_of_book')] = 2024 - books['published_year']
```

```python
columns_of_interest = ['num_pages', 'age_of_book', 'missing description', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method='spearman')

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
book_missing = books[~(books['description'].isna()) &
                     ~(books['num_pages'].isna()) &
                     ~(books['average_rating'].isna()) &
                     ~(books['published_year'].isna())
]
```

```python
book_missing
```

```python
book_missing["categories"].value_counts().reset_index().sort_values("count",ascending=False)
```

```python
book_missing['words_in_description'] = book_missing['description'].str.split().str.len()
```

```python
book_missing
```

```python
book_missing.loc[book_missing['words_in_description'].between(1,4),'description'] = 1000
```

```python
book_missing_25_words = book_missing[book_missing['words_in_description'] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words['title_and_subtitle'] = (
    np.where(book_missing_25_words['subtitle'].isna(), book_missing_25_words['title'],
             book_missing_25_words[['title','subtitle']].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[['isbn13','description']].astype(str).agg(" ".join, axis=1)
```

```python
(
    book_missing_25_words
    .drop(["subtitle","missing description","age_of_book","words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
